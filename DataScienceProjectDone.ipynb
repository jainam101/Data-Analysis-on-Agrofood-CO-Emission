Visualizing Top 10 Emitting Areas

```python
df = pd.read_csv('Agrofood_co2_emission.csv')

df.columns = df.columns.str.strip()

print("Dataset Info:")
print(df.info())

arr = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(arr))

area_emission = df.groupby('Area')['total_emission'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
area_emission.head(10).plot(kind='bar', color='green')
plt.title("Top 10 Areas by Total Emissions")
plt.ylabel("Total Emissions")
plt.xticks(rotation=45)
plt.show()
```

Different Types of charts and graphs

```python
# Discrete Frequency: Count of years
year_freq = df['Year'].value_counts().sort_index()
print("Year Frequency:\n", year_freq)

# Continuous Frequency: total_emission
bins = [0, 10000, 50000, 100000, 200000, df['total_emission'].max()]
df['EmissionBins'] = pd.cut(df['total_emission'], bins)
print("Emission Bins:\n", df['EmissionBins'].value_counts())

# Bar Diagram
year_freq.plot(kind='bar', title='Year-wise Data Count')
plt.show()

# Histogram
plt.hist(df['total_emission'], bins=20, color='skyblue')
plt.title("Histogram of Total Emission")
plt.xlabel("Emission")
plt.ylabel("Frequency")
plt.show()

# Pie Chart
top_areas = df.groupby('Area')['total_emission'].sum().nlargest(5)
top_areas.plot.pie(autopct='%1.1f%%', title="Top 5 Areas by Emission")
plt.ylabel("")
plt.show()

# Frequency Curve (KDE)
sns.histplot(df['total_emission'], kde=True)
plt.title("Frequency Curve of Total Emission")
plt.show()
```

```python
print("Mean:", df['total_emission'].mean())
print("Median:", df['total_emission'].median())
print("Mode:", df['total_emission'].mode()[0])
print("Std Dev:", df['total_emission'].std())
print("Variance:", df['total_emission'].var())
```

```python
# Clean column names - No need to strip again as it was likely done before
# df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace('\u00a0', ' ', regex=True)

# Rename target column if needed
# We need to use the actual column name (likely 'Average Temperature °C')
if 'Average Temperature °C' in df.columns: # Changed to likely actual column name
    df.rename(columns={'Average Temperature °C': 'AvgTemp'}, inplace=True) # Changed to likely actual column name

# Confirm it's renamed
print(df.columns.tolist())

# Convert to numeric
# Use updated column name (if renamed) or original (if not)
if 'AvgTemp' in df.columns:
    df['AvgTemp'] = pd.to_numeric(df['AvgTemp'], errors='coerce')
else:
    df['Average Temperature °C'] = pd.to_numeric(df['Average Temperature °C'], errors='coerce') # Assuming this is the original
df['total_emission'] = pd.to_numeric(df['total_emission'], errors='coerce')

# Drop missing rows
df = df.dropna(subset=['AvgTemp' if 'AvgTemp' in df.columns else 'Average Temperature °C', 'total_emission'])

# Define features and label
X = df[['AvgTemp' if 'AvgTemp' in df.columns else 'Average Temperature °C']] # Use correct column
y = df['total_emission']

# Split and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
lr = LinearRegression()
lr.fit(X_train, y_train)

# Results
print("Coefficient:", lr.coef_)
print("Intercept:", lr.intercept_)
print("R^2 Score:", lr.score(X_test, y_test))
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
print("KNN R^2 Score:", knn.score(X_test, y_test))
```

```python
features = df[['total_emission', 'Rural population', 'Urban population']].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualize
plt.figure(figsize=(8, 5))
plt.scatter(df['Urban population'], df['total_emission'], c=df['Cluster'], cmap='rainbow')
plt.xlabel("Urban Population")
plt.ylabel("Total Emission")
plt.title("KMeans Clustering of Emissions")
plt.show()
```
